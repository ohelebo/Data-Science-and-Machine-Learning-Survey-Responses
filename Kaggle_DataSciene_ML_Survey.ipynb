```python
#Create a copy of the dataframe so i can easily rollback 
#incase there is a mistake instead of imprting the entire dataset (which is heavy)
```

```python
data_copy = data
```

```python
#Viewing the column names so I can drop the unneccesary columns easily
data_copy.columns
```

```python
#Remove unwanted columns as we won't be needing them for the analysis
unwanted_column = ["Time from Start to Finish (seconds)"]
unwanted_column
```

```python
data_copy = data_copy.drop(columns = unwanted_column)
data_copy
```

```python
id_vars = list(data_copy.columns[:7])
id_vars
```

```python
data_questions = list(data_copy.columns[7:])
data_questions
```

```python
#Unpivot the table using the melt function
data_copy_melted = data_copy.melt(id_vars = id_vars, value_vars = data_questions, var_name = "Question+Subquestion", value_name = "Answer")
data_copy_melted
```

```python
ques_data = pd.read_excel(pwd + "\\Data_kaggle_survey_2021_responses_copyy.xlsx", sheet_name = "Questions")
```

```python
ques_data
```

```python
#Drop unwanted column(s)
#make a duplicate of the data
questions = ques_data.copy()
```

```python
questions.drop(columns = "Subquestion", inplace = True)
```

```python
questions
```

```python
#Merge both pivoted (melted df) and questions df using the join statement
alldata = pd.merge(data_copy_melted, questions, how='left', on= "Question+Subquestion")
```

```python
alldata
```

```python
alldata.head(20)
```

```python
#Count the number of responses to particular questions by grouping by question and counting by the Answer column
respondents = alldata.groupby("Question")["Answer"].count().reset_index()
respondents.rename(columns = {"Answer": "No. of respondents"}, inplace = True)
```

```python
respondents
```

```python
#Merge respondents to alldata
newdata = pd.merge(alldata,respondents, how = "left", on = "Question")
newdata
```

```python
dataC = newdata.copy()
dataC["Q4"] = dataC["Q4"].apply(lambda x: re.sub(r"[^a-zA-Z0-9\s]+", "'", x))
```
